```python
import pandas as pd
df = pd.read_csv("data.csv")
df
```

```python
ContinousDataFrame=df.select_dtypes(exclude=['object','category'])
ContinousDataFrame
```

```python
CategoricalDataFrame=df.select_dtypes(include=['object','category'])
CategoricalDataFrame
```

```python
from sklearn.preprocessing import OneHotEncoder
le = OneHotEncoder(sparse=False)
```

```python
CategoricalDataFrame = pd.DataFrame(le.fit_transform(CategoricalDataFrame))
CategoricalDataFrame
```

```python
Final = pd.concat([ContinousDataFrame,CategoricalDataFrame],axis=1)
Final
```

```python
X = Final.drop(["price"],axis=1)
X
```

```python
y = df.price
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
#Note random_state is the pattern for spliting
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
prediction=pd.DataFrame(clf.predict(X_test))
prediction
```

```python
y_test
```

```python
#Accuracy
clf.score(X_test,y_test)
```
